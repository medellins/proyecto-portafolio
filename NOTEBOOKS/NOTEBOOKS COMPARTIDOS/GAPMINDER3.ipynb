```python
import plotly.express as ply
import numpy as np
import pandas as pd 
```

```python
gdp_data=pd.read_csv("../input/gapminder-datacamp-2007/gapminder_full.csv")
```

```python
gdp_data.head()
```

```python
gdp_data["gdp"]=gdp_data["gdp_cap"]*gdp_data["population"]
```

```python
gdp_data.groupby("year")["gdp_cap"].describe()
```

```python
gdp_data.groupby("year")["gdp"].describe()
```

**Population of country over life expectancy and GDP per capita by year**

```python
ply.scatter(data_frame= gdp_data,
           x= "gdp_cap",
           y="life_exp",
           animation_frame="year",
           animation_group="country",
           size="population", 
           color="continent", 
           hover_name="country",
           log_x=False, size_max=55, range_x=[298,114000], range_y=[25,100],
           hover_data=["gdp"] 
           )
```
