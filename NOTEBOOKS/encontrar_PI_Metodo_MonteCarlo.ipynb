```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Simular una cantidad de n puntos en el cuadrado por eje
n=3000
eje_x=np.random.uniform(-1,1, size=n)
eje_y=np.random.uniform(-1,1, size=n)
```

```python
#imprimir los puntos aleatorios
print(eje_x[0:5])
print(eje_y[0:5])#cada punto x con y estar√≠a ubicado en esa coordenada
```

```python
#Graficar los puntos
plt.gca().add_patch(plt.Circle((0,0), radius=1))
plt.gca().set_aspect('equal', adjustable='box')#esto hace que el aspecto sea igual vertical/horizontalmente
plt.plot(eje_x, eje_y, '.', color='orange')
plt.show()
```

```python
#contando los puntos dentro del c√≠rculo
dist_origen=np.sqrt(eje_x*eje_x+eje_y*eje_y)
suma_interior=np.sum(dist_origen<1)
```

```python
print(suma_interior)
```

```python
#el resultado de suma interior hay que dividirlo entre todos los puntos,ver si el resultado se aproxima al valor de PI
4*suma_interior/n
```

```python
#vemos que el resultado fue 3.6 lo que a√∫n esta alejado del valor de PI
#ahora tendremos que ir aumentando el numero de puntos o el valor de "n" hasta tener un valor muy cercano a PI
```
